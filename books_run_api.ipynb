```python
#pull in our csv file
df_ebay =pd.read_csv('isbn_df_2')
```

```python
df_ebay
```

```python
#data cleaning
df_ebay= df_ebay.dropna()
# df_ebay = df_ebay.reset_index()
df_ebay = df_ebay[df_ebay['isbn'].apply(lambda x: x.isnumeric())]
# df_ebay['isbn']=df_ebay['ISBN'].astype(int)
```

```python
df_ebay
```

```python
#calling all the ISBNs via API and then appending to a list
data_list=[]
for i in df_ebay['isbn']:
    url = 'https://booksrun.com/api/price/sell/'+str(i)+'?key=so3w3ptkzjfpb3oxjs6z'
    r = requests.get(url)
    data = r.json()
    data_list.append(data)
#     df_1= pd.DataFrame(data['result'],index=[0])
#     df_1=df_1[['text']].T
#     print(df_1)
#     df=pd.concat([df, df_1])
```

```python
data_list
```

```python
#putting the list into a DataFrame
df_data_list = pd.json_normalize(data_list)
```

```python
df_data_list
```

```python
#concat DataFrames
df_full = pd.concat([df_data_list,df_ebay],axis =1)
```

```python
df_full
```

```python
#data cleaning and changing numbers 
df_full['price_list'] = df_full['prices'].replace('[\$,]', '', regex=True).astype(float)
df_full['result.text.Average']=df_full['result.text.Average'].astype(float)
df_full['result.text.Good']=df_full['result.text.Average'].astype(float)
df_full['result.text.New']=df_full['result.text.Average'].astype(float)
```

```python
#adding the difference column
df_full['difference']=df_full['price_list']-df_full['result.text.Good']
```

```python
#sorting values
df_full.sort_values(by=['difference'])
```

```python
#putting into a csv 
df_full.to_csv('final_list.csv')
```
